```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
gender = ['Male', 'Female', 'Male', 'Male', 'Female']
df = pd.DataFrame({'gender': gender})

# Step 2: Plot
sns.countplot(x="gender", data=df)
plt.title("Gender Count")
plt.show()
```

```python
# Q1.
# Line Plot:
# Plot the monthly temperature for a year using a line chart. Add title, axis labels, and grid.

months = ["Jan", "Feb" , "Mar", "Apr", "May", "June", "July", "Aug", "Sem", "Oct", "Nov", "Dec"] 
temp = [10, 15, 20, 30, 45, 43, 50, 51, 50, 45, 35, 40]

sns.lineplot(x = temp, y = months)
plt.xlabel("temp")
plt.ylabel("Months")
plt.grid(True)
```

```python
#Q2. Show the total sales of 5 products in a bar plot. Add value labels on top of each bar.

products = ["Apple", "Phone", "Laptop", "Biryani", "Chapli Kabab"]
sales = [5, 6, 3, 5, 6]

ax = sns.barplot(x = products,  y = sales)

for i, value in enumerate(sales):
    ax.text(i, value + 0.1, str(value))

plt.title("Total Sales of 5 Products")
plt.ylabel("Sales")
plt.xlabel("Products")
plt.show()

# ax.text()
```

```python
# Q3. Generate a list of 100 random ages (1860) and plot their distribution using a histogram with 8 bins.

ages = []

for i in range(0, 100):
    ages.append(random.randint(18, 60))

# print(ages)

sns.histplot(ages, bins=8)
plt.show()
```

```python
# Q4. Simulate data for "Study Hours" vs "Exam Scores" and plot a scatter plot to visualize correlation.

study_hour = [8, 10, 4, 5, 9, 8, 10, 3, 5, 4]
scores = [80, 51, 90, 50, 99, 41, 98, 55, 8, 90]

sns.scatterplot(x = study_hour, y = scores)
plt.xlabel("Study Hour")
plt.ylabel("Scores")
plt.show()
```

```python
# Q6. Plot a pie chart showing percentage of devices used to visit your website: Mobile, Desktop, Tablet.
labels = ["Mobile", "Desktop", "Tablet"]
val = [50, 40, 10]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.pie(val, labels = labels, autopct='%1.1f%%', startangle=361, colors= colors)
# plt.tight_layout()
plt.show()
```

```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

# Create the figure
plt.figure(figsize=(10, 6))

# Plot with custom styling
plt.plot(x, y, linewidth=3, marker='o', markersize=6, label='Sine wave', color='#1f77b4')
plt.plot(x, z, linewidth=3, marker='s', markersize=6, label='Cosine wave', color='#ff7f0e')

# Styling labels, title, ticks
plt.xlabel('Time (s)', fontsize=14)
plt.ylabel('Amplitude', fontsize=14)
plt.title('Dark Themed Plot with Custom Styles', fontsize=16, pad=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Customize grid for dark background
plt.grid(color='white', linestyle='--', alpha=0.3)

# Legend with styling
plt.legend(fontsize=12, frameon=True, facecolor='#333333', edgecolor='white')

# Optional: tighten layout
# plt.tight_layout()

plt.show()
```
